# **Separating Features and Target Variable**

```python
features = data.columns.drop('Y')
features
```

```python
train_size = int(0.8 * len(data))

train_data = data.iloc[:train_size]
test_data = data.iloc[train_size:]

X_train = train_data[features]
y_train = train_data['Y']
X_test = test_data[features]
y_test = test_data['Y']
```

# **Scaling to Standardize the features**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# **Principal Component Analysis for dimensionality reduction**

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=7)  # Retain 95% variance
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

# **Logistic Regression with Anomaly Detection (Isolation Forest)**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import IsolationForest

isolation_forest = IsolationForest(contamination=0.05)  # Adjust contamination as needed
y_anomaly = isolation_forest.fit_predict(X_train_pca)

X_train_clean = X_train_pca[y_anomaly == 1]
y_train_clean = y_train[y_anomaly == 1]
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train_clean, y_train_clean)

y_pred = logistic_regression.predict(X_test_pca)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

### **Saving the logistic regression model**

```python
import joblib
joblib.dump(logistic_regression, 'logistic_regression_model.pkl')
```

# **Anomaly Detection with K-Nearest Neighbours**

```python
from sklearn.neighbors import NearestNeighbors

num_neighbors = 3
knn = NearestNeighbors(n_neighbors=num_neighbors).fit(X_train_pca)

# Calculate distances and indices for training data
distances_train, indices_train = knn.kneighbors(X_train_pca)

# Calculate average distance for training data
average_distance_train = np.mean(distances_train, axis=1)

# Define threshold based on training data
threshold = np.percentile(average_distance_train, 99)  # top 5% anomalies

# Identify anomalies in training data
anomalies_train = average_distance_train > threshold

# Create target variable for training data
y_train_anomalous = anomalies_train.astype(int)

log_reg = LogisticRegression()
log_reg.fit(X_train_pca, y_train_anomalous)

distances_test, indices_test = knn.kneighbors(X_test_pca)
average_distance_test = np.mean(distances_test, axis=1)
anomalies_test = average_distance_test > threshold

y_pred = anomalies_test.astype(int)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# **Anomaly Detection with One-Class SVM**

```python
from sklearn.svm import OneClassSVM
from sklearn.metrics import roc_auc_score
one_class_svm = OneClassSVM(nu=0.01, kernel='rbf')
one_class_svm.fit(X_train_pca)

y_pred = one_class_svm.predict(X_test_pca)
y_pred = np.where(y_pred == -1, 1, 0)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

roc_auc = roc_auc_score(y_test, y_pred)
print(f"ROC AUC: {roc_auc}")
```

# **Anomaly Detection with Neural Network**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


X = data[['BDIY', 'DXY', 'VIX', 'USGG30YR', 'GT10', 'GTDEM30Y']]
y = data['Y']

scaled = StandardScaler()
X_scaled = scaled.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)

model = Sequential()

model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=15, batch_size=16, validation_split=0.2, verbose=1)

y_pred = (model.predict(X_test) > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

plt.figure(figsize=(12, 6))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left')

plt.tight_layout()
plt.show()
```

# **Backtesting**

```python
import yfinance as yf
# Download data from Yahoo Finance
tickers = ['TRI', 'DX-Y.NYB', 'JPY=X', 'GBPUSD=X', 'CL=F', '^VIX', '^TYX', '2YY=F', 'TBIL', 'MSCI']
yf_data = yf.download(tickers, start='2021-01-01', end='2023-12-31')['Close']

# Check for available tickers in the data
available_tickers = [ticker for ticker in tickers if ticker in yf_data.columns]
print("Available tickers in data:", available_tickers)

# If any tickers are missing, we'll add them with NaN values (to avoid KeyErrors)
missing_tickers = set(tickers) - set(available_tickers)
if missing_tickers:
    print(f"Missing tickers: {missing_tickers}")
    for ticker in missing_tickers:
        yf_data[ticker] = np.nan  # Add missing tickers with NaN values

# Clean up any rows where all tickers are NaN
yf_data = yf_data.dropna(how='all')

# Check for NaN values and fill them if necessary
print("NaN values in yf_data after cleaning:", yf_data.isna().sum())

# Fill missing values with forward fill and backward fill
yf_data = yf_data.ffill().bfill()

# Now proceed with scaling and PCA
scaler = StandardScaler()
yf_scaled = scaler.fit_transform(yf_data)

# Apply PCA
# pca = PCA(n_components=7)
yf_pca = pca.fit_transform(yf_scaled)
yf_pca_df = pd.DataFrame(yf_pca, index=yf_data.index, columns=[f"PC{i+1}" for i in range(7)])

# Proceed with model prediction and backtesting...
# Predict anomalies
yf_pca_df['Anomaly'] = logistic_regression.predict(yf_pca_df)

# Generate signals (Sell if anomaly, otherwise hold)
yf_pca_df['Signal'] = np.where(yf_pca_df['Anomaly'] == 1, -1, 0)

# Calculate daily returns for TRI
asset_returns = yf_data['^VIX'].pct_change().dropna()

# Align signal and returns
yf_pca_df = yf_pca_df.iloc[1:]  # Align with asset_returns after pct_change
yf_pca_df['Strategy Returns'] = asset_returns.values * yf_pca_df['Signal'].values

# Calculate cumulative returns
yf_pca_df['Cumulative Strategy Returns'] = (1 + yf_pca_df['Strategy Returns']).cumprod()
yf_pca_df['Cumulative Asset Returns'] = (1 + asset_returns).cumprod()

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(yf_pca_df['Cumulative Strategy Returns'], label='Anomaly Detection Strategy', color='blue')
plt.plot(yf_pca_df['Cumulative Asset Returns'], label='Buy & Hold Returns', color='orange')
plt.title('Backtest: Anomaly Detection Strategy vs. Buy & Hold')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.grid()
plt.show()

# Performance metrics
sharpe_ratio = (yf_pca_df['Strategy Returns'].mean() / yf_pca_df['Strategy Returns'].std()) * np.sqrt(252) if yf_pca_df['Strategy Returns'].std() > 0 else 0
total_return = yf_pca_df['Cumulative Strategy Returns'].iloc[-1] - 1
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Total Return: {total_return:.2%}")
```

# **Monitoring the Market using Yahoo Finance API**

```python
import discord
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from google.colab import userdata
import asyncio
import time

import google.generativeai as genai
genai.configure(api_key=userdata.get("GOOGLE_API_KEY"))

DISCORD_BOT_TOKEN = userdata.get('DISCORD_KEY')
CHANNEL_ID = int(userdata.get('CHANNEL_ID'))

intents = discord.Intents.default()
client = discord.Client(intents=intents)

class MarketMonitor:
    def __init__(self, tickers, portfolio_weights, model, scaler, pca, lookback_period='3m', interval='1d', discord_token='', channel_id=None):
        self.tickers = tickers
        self.portfolio_weights = portfolio_weights
        self.model = model
        self.scaler = scaler
        self.pca = pca
        self.lookback_period = lookback_period
        self.interval = interval
        self.discord_token = discord_token
        self.channel_id = channel_id
        self.client = discord.Client(intents=discord.Intents.default())

        @self.client.event
        async def on_ready():
            print(f'Logged in as {self.client.user}')

    async def send_discord_message(self, message):
        if self.channel_id:
            channel = self.client.get_channel(self.channel_id)
            if channel:
                await channel.send(message)
            else:
                print(f"Invalid channel ID: {self.channel_id}")
        else:
            print("No channel ID provided for alerts.")

    def fetch_latest_data(self):
        data = {}
        for ticker in self.tickers:
            try:
                stock = yf.Ticker(ticker)
                hist = stock.history(period='1d', interval='1m')
                if not hist.empty:
                    data[ticker] = hist['Close'].iloc[-1]
                    print(f"Fetched {ticker}: {data[ticker]}")
                else:
                    print(f"No data available for {ticker}")
            except Exception as e:
                print(f"Error fetching {ticker}: {e}")
                data[ticker] = None
        return data

    def prepare_features(self, current_data):
        features = pd.DataFrame([current_data])
        for ticker in self.tickers:
            if ticker not in features.columns:
                features[ticker] = 0
        features = features[self.tickers]
        return features

    def detect_anomaly(self, features):
        try:
            features_scaled = self.scaler.transform(features)
            features_pca = self.pca.transform(features_scaled)
            prediction = self.model.predict(features_pca)
            probability = self.model.predict_proba(features_pca)
            return prediction[0], probability[0]
        except Exception as e:
            print(f"Error in anomaly detection: {e}")
            return None, None

    async def get_investment_strategy(self, current_data, portfolio_impacts):
        user_prompt = f"""
        Current Market Data:
        {current_data}

        Portfolio Impacts:
        {portfolio_impacts}

        Based on the above data, provide an investment strategy for the portfolio.
        Suggest if the user should hold, buy, or sell, and provide the rationale behind your decision.
        """

        system_prompt = "You are a financial advisor with expertise in investment strategies. Your goal is to analyze market data and portfolio impacts and suggest an optimal investment strategy."

        try:
            model = genai.GenerativeModel(model_name="gemini-1.5-flash")
            response = model.generate_content(system_prompt + " " + user_prompt)
            return response.result.strip()
        except Exception as e:
            print(f"Error generating strategy with Gemini LLM: {e}")
            return "Unable to generate strategy at this time."

    async def monitor_market_and_alert(self, channel, interval_seconds=60):
        previous_data = None

        while True:
            try:
                print(f"\nChecking market conditions at {datetime.now()}")

                current_data = self.fetch_latest_data()
                features = self.prepare_features(current_data)

                is_anomaly, probability = self.detect_anomaly(features)

                if is_anomaly is not None:
                    message = f"Anomaly detected: {is_anomaly}\nProbability: {probability}\n"

                    if previous_data is not None:
                        impacts, total_impact = self.calculate_portfolio_impact(current_data, previous_data)
                        message += "\nPortfolio Impacts:\n"
                        for ticker, impact in impacts.items():
                            message += f"{ticker}: {impact*100:.2f}%\n"
                        message += f"Total Portfolio Impact: {total_impact*100:.2f}%\n"

                    # Get investment strategy from Gemini LLM
                    strategy = await self.get_investment_strategy(current_data, impacts)
                    message += f"\nInvestment Strategy: {strategy}"

                    # Send alert to Discord channel
                    await channel.send(message)

                previous_data = current_data

            except Exception as e:
                print(f"Error in market monitoring: {e}")

            await asyncio.sleep(interval_seconds)

@client.event
async def on_ready():
    print(f"We have logged in as {client.user}")

    channel = client.get_channel(CHANNEL_ID)
    if channel is None:
        print("Channel not found. Check the CHANNEL_ID.")
        return

    monitor = MarketMonitor(
        tickers=tickers,
        portfolio_weights=portfolio_weights,
        model=logistic_regression,  # Your trained model
        scaler=scaler,  # Your scaler
        pca=pca,  # Your PCA
        discord_token=DISCORD_BOT_TOKEN,
        channel_id=CHANNEL_ID
    )

    await monitor.monitor_market_and_alert(channel, interval_seconds=60)
```
